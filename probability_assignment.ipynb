```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Question 1

heads = 0
tails = 0

for i in range(10000):
    toss = random.choice(['H', 'T'])

    if toss == 'H':
        heads += 1
    else:
        tails += 1

print("Probability of Heads:", heads / 10000)
print("Probability of Tails:", tails / 10000)

count_sum_7 = 0
trials = 10000

for i in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)

    if die1 + die2 == 7:
        count_sum_7 += 1

print("Probability of Sum 7:", count_sum_7 / trials)

# Question 2

def estimate_probability(trials=10000):
    success = 0

    for i in range(trials):
        found_six = False

        for j in range(10):
            if random.randint(1, 6) == 6:
                found_six = True
                break

        if found_six:
            success += 1

    return success / trials

print("Estimated Probability:", estimate_probability())

# Question 3

balls = ['Red'] * 5 + ['Green'] * 7 + ['Blue'] * 8

previous_blue = 0
red_after_blue = 0

previous = random.choice(balls)

for i in range(1000):
    current = random.choice(balls)

    if previous == 'Blue':
        previous_blue += 1

        if current == 'Red':
            red_after_blue += 1

    previous = current

conditional_probability = red_after_blue / previous_blue

print("P(Red | Previous Blue):", conditional_probability)

p_red = 5 / 20
p_blue = 8 / 20
p_blue_given_red = p_blue

bayes_result = (p_blue_given_red * p_red) / p_blue

print("Bayes Theorem Result:", bayes_result)

# Question 4

values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]

sample = np.random.choice(values, size=1000, p=probabilities)

print("Empirical Mean:", np.mean(sample))
print("Variance:", np.var(sample))
print("Standard Deviation:", np.std(sample))

# Question 5

data = np.random.exponential(scale=5, size=2000)

plt.hist(data, bins=30, density=True)

x = np.linspace(0, max(data), 1000)
pdf = (1/5) * np.exp(-x/5)

plt.plot(x, pdf)

plt.title("Exponential Distribution")
plt.xlabel("Value")
plt.ylabel("Density")

plt.show()

# Question 6

uniform_data = np.random.uniform(0, 1, 10000)

sample_means = []

for i in range(1000):
    sample = np.random.choice(uniform_data, size=30)
    sample_means.append(np.mean(sample))

plt.hist(uniform_data, bins=30)
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

plt.hist(sample_means, bins=30)
plt.title("Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
